## Data loading

### Subtask:
Load the dataset "Self-Esteem Questionnaire for Teacher Trainees  (Responses).xlsx" into a pandas DataFrame.


**Reasoning**:
Load the excel file into a pandas DataFrame and display the first few rows.



```python
import pandas as pd

try:
    df = pd.read_excel('Self-Esteem Questionnaire for Teacher Trainees  (Responses).xlsx')
    display(df.head())
except FileNotFoundError:
    print("Error: File not found. Please ensure the file 'Self-Esteem Questionnaire for Teacher Trainees  (Responses).xlsx' is in the current directory or provide the correct file path.")
    df = None # Assign None to df in case of error
except Exception as e:
    print(f"An error occurred: {e}")
    df = None
```

## Data exploration

### Subtask:
Explore the dataset to understand its structure and characteristics.


**Reasoning**:
I need to explore the dataset by examining its shape, data types, missing values, and descriptive statistics.



```python
# Examine the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Check data types and identify potential inconsistencies
print("\nData Types:")
df.info()

# Identify missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
numerical_cols = df.select_dtypes(include=['number']).columns
display(df[numerical_cols].describe())

# Preliminary observations
print("\nPreliminary Observations:")
print("The dataset has", df.shape[0], "rows and", df.shape[1], "columns.")
print("The data types mostly seem correct, with some numerical columns representing questionnaire responses and age. Timestamp also seems to be correctly inferred as datetime64.")

# Check for duplicate rows
duplicate_rows = df[df.duplicated()]
if not duplicate_rows.empty:
    print("\nDuplicate Rows:")
    display(duplicate_rows)
else:
    print("\nNo duplicate rows found.")
```

## Data analysis

### Subtask:
Analyze the distribution of responses for each question in the questionnaire.


**Reasoning**:
Calculate the mean, median, and standard deviation for each question and generate a textual description.  Compile the results into a summary dictionary.



```python
import pandas as pd
summary = {}
question_columns = df.select_dtypes(include=['number']).columns
question_columns = question_columns.drop('Age') #exclude the Age column
for col in question_columns:
    mean = df[col].mean()
    median = df[col].median()
    std = df[col].std()

    description = f"Question '{col}' has a mean of {mean:.2f}, median of {median:.2f}, and standard deviation of {std:.2f}."

    if mean > median:
        description += " The responses are slightly right-skewed."
    elif mean < median:
        description += " The responses are slightly left-skewed."
    else:
        description += " The responses are approximately symmetric."

    summary[col] = {
        'mean': mean,
        'median': median,
        'std': std,
        'description': description
    }

for key, value in summary.items():
    print(value['description'])
```

## Data visualization

### Subtask:
Visualize the distribution of responses for each question in the questionnaire using histograms.


**Reasoning**:
Generate the histograms for each question, excluding irrelevant columns.



```python
import matplotlib.pyplot as plt

# Select numerical columns, excluding 'Timestamp', 'Age', and 'Signature'
numerical_cols = df.select_dtypes(include=['number']).columns
numerical_cols = numerical_cols.drop(['Age'])

# Number of histograms
num_plots = len(numerical_cols)

# Calculate the optimal number of rows and columns for subplots
ncols = 3  # Adjust the number of columns as needed
nrows = (num_plots + ncols - 1) // ncols

# Create subplots
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 5 * nrows))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Iterate through numerical columns and create histograms
for i, col in enumerate(numerical_cols):
    axes[i].hist(df[col], bins=10, color='skyblue', edgecolor='black')  # Adjust bins as needed
    axes[i].set_title(f'Distribution of Responses for Question: {col}', fontsize=10)
    axes[i].set_xlabel("Response Value")
    axes[i].set_ylabel("Frequency")

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()


import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Assuming 'df' is your DataFrame
# Select only numerical columns for correlation analysis, excluding 'Signature' and other non-numerical columns
numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(12, 10))  # Adjust the figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Questionnaire Responses")
plt.show()
# Identify highly correlated questions (e.g., absolute correlation > 0.7)
highly_correlated = correlation_matrix[correlation_matrix.abs() > 0.7]
highly_correlated = highly_correlated.unstack().sort_values(ascending=False)
highly_correlated = highly_correlated[highly_correlated != 1]  # Remove self-correlation

print("\nHighly Correlated Questions (absolute correlation > 0.7):")
print(highly_correlated)



# Assuming 'df' is your DataFrame
# Select only numerical columns for correlation analysis
numerical_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Display the correlation matrix
print(correlation_matrix)
```
